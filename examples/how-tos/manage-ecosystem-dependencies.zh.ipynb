# 如何安装和管理依赖项

LangGraph.js 是 [LangChain](https://js.langchain.com/) 生态系统的一部分，
其中包括初级
[`langchain`](https://www.npmjs.com/package/langchain) 包以及
包含与各个第三方提供商集成的软件包。他们
可以具体为
[`@langchain/anthropic`](https://www.npmjs.com/package/@langchain/anthropic),
其中包含仅针对人择聊天模型的集成，或者广泛到
[`@langchain/community`](https://www.npmjs.com/package/@langchain/community),
其中包含更广泛的社区贡献的集成。

这些包，以及 LangGraph.js 本身，都依赖于
[`@langchain/core`](https://www.npmjs.com/package/@langchain/core)，其中
包含这些包扩展的基本抽象。

确保所有集成及其类型相互交互
正确地说，重要的是他们都使用相同版本的
[[代码4]]。安装 LangGraph 时，您还应该安装 `@langchain/core`：
```bash
$ npm install @langchain/langgraph @langchain/core
```
`@langchain/core` 必须单独安装，因为它是 `@langchain/langgraph` 的对等依赖项。
这是为了帮助包管理器解析 `@langchain/core` 的单个版本。

尽管如此，在某些情况下，您的包管理器可能会解析多个版本的 core，这可能会导致意外的 TypeScript 错误或其他奇怪的行为。如果您需要保证只有一个版本的 `@langchain/core` 是添加 `"resolutions"` 或
项目的 `package.json` 中的 `"overrides"` 字段。具体字段名称取决于您的包管理器。以下是一些示例：

<div class="警告提示">
<p class="admonition-title">提示</p>
<p>
<code>yarn</code> 或 <code>pnpm</code> 的 <code>resolutions</code> 或 <code>pnpm.overrides</code> 字段必须在根 <code>package.json</code> 文件中设置。

另请注意，我们为分辨率指定了确切的版本。
</p>
</div>

如果您使用 `yarn`，则应设置 [`"resolutions"`](https://yarnpkg.com/cli/set/resolution)：
```json
{
  "name": "your-project",
  "version": "0.0.0",
  "private": true,
  "engines": {
    "node": ">=18"
  },
  "dependencies": {
    "@langchain/anthropic": "^0.2.15",
    "@langchain/langgraph": "^0.2.0"
  },
  "resolutions": {
    "@langchain/core": "0.2.31"
  }
}
```
对于 `npm`，使用 [`"overrides"`](https://docs.npmjs.com/cli/v10/configuring-npm/package-json#overrides)：
```json
{
  "name": "your-project",
  "version": "0.0.0",
  "private": true,
  "engines": {
    "node": ">=18"
  },
  "dependencies": {
    "@langchain/anthropic": "^0.2.15",
    "@langchain/langgraph": "^0.2.0"
  },
  "overrides": {
    "@langchain/core": "0.2.31"
  }
}
```
对于 `pnpm`，使用嵌套的 [`"pnpm.overrides"`](https://pnpm.io/package_json#pnpmoverrides) 字段：
```json
{
  "name": "your-project",
  "version": "0.0.0",
  "private": true,
  "engines": {
    "node": ">=18"
  },
  "dependencies": {
    "@langchain/anthropic": "^0.2.15",
    "@langchain/langgraph": "^0.2.0"
  },
  "pnpm": {
    "overrides": {
      "@langchain/core": "0.2.31"
    }
  }
}
```
